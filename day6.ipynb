# Class Inheritance

```python
class Person:
    def __init__(self, age: float, height: int, nationality: str, color: str='black'):
        assert age >= 0, 'อายุมากกว่าหรือเท่ากับ 0'
        self.age = age
        self.height = height
        self.nationality = nationality
        self.color = color
    
    def speak(self, message: str) -> None:
        print(message)
```

```python
namo = Person(0.1, 0.1, 'Indian')
namo.color
```

```python
class Muslum(Person):
    def __init__(self, capacity: float, **kwargs):
        super().__init__(**kwargs)
        self.capacity = capacity
        self.is_equip_weapon = False
        self.n_weapon = 1
        
    def equip_weapon(self, weapon: str, ammo: int):
        self.capacity -= ammo
        self.is_equip_weapon = True
    
    def unequip_weapon(self):
        self.is_equip_weapon = False
```

```python
kanplu = Muslum(age=14, height=160, nationality='Iraq', capacity=100)
```

```python
kanplu.equip_weapon('AK-47', 80)
```

```python
kanplu.is_equip_weapon
```

```python
kanplu.capacity
kanplu.n_weapon
```

```python
class AlQaedi(Muslum):
    def __init__(self, n_weapon: int, **kwargs):
        super().__init__(**kwargs)
        self.n_weapon = n_weapon
        self.is_bomb_planted = False
        
    def plant_bomb(self):
        self.is_bomb_planted = True
    
    def detonate_bomb(self):
        if self.is_bomb_planted:
            self.speak('!!!!!!!!!!!!!!!!!!')
            import random
            is_survive = random.choice([True, False])
            if not is_survive:
                del self
        else:
            self.speak('Please plant a bomb first!')
```

```python
namo = AlQaedi(age=14, height=160, nationality='Iraq', capacity=100, n_weapon=9999)
```

```python
namo.plant_bomb()
namo.detonate_bomb()
```
