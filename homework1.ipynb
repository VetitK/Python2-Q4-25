# Homework 1: Rectangula Coordinates

$p V q \equiv ~p \wedge ~q$

```python
from math import sqrt, atan, degrees

class RectangularCoordinates:
    def __init__(self, x: int, y: int) -> None:
        assert isinstance(x, int) and isinstance(y, int), "Pls enter int plssssss"
        self.x = x
        self.y = y
    
    def get_coordinate(self) -> tuple[int, int]:
        return self.x, self.y
    
    def is_on_axis(self) -> bool:
        # if self.x == 0 or self.y == 0:
        #     return True
        # else:
        #     return False
        
        return self.x == 0 or self.y == 0
    
    def get_quadrant(self) -> int:
        if self.x > 0 and self.y > 0:
            return 1
        elif self.x < 0 and self.y > 0:
            return 2
        elif self.x < 0 and self.y < 0:
            return 3
        elif self.x > 0 and self.y < 0:
            return 4
        else:
            print("Not on any q")
            return 0
    
    def get_distance_from_x_axis(self) -> int:
        return abs(self.y)
        
    def get_distance_from_y_axis(self) -> int:
        return abs(self.x)
    
    def get_distance_origin(self) -> float:
        return sqrt(self.x ** 2 + self.y ** 2)

    def get_midpoint(self, a: int, b: int) -> tuple[float, float]:
        return (self.x + a)/2, (self.y + b)/2
    
    def get_distace(self, a: int, b: int) -> float:
        return sqrt((self.x - a) ** 2 + (self.y - b) ** 2)
    
    def get_angle_with_x_axis(self) -> float:
        return degrees(atan(self.y / self.x))
    
    def __repr__(self):
        return f"RectangularCoordinates(x={self.x}, y={self.y})"
```

```python
p = RectangularCoordinates(-4, -3)
```
