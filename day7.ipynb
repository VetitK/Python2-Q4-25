## เขียน File

```python
f = open('เราจะทำตามสัญญา.txt', 'w+')
```

```python
uncle2 = """วันที่ชาติและองค์ราชา มวลประชาอยู่มาพ้นภัย
ขอดูแลคุ้มครองด้วยใจ นี่คือคำสัญญา
"""
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
f.write(uncle2)
```

```python
f.flush()
```

```python
lines = ['วันที่ชาติและองค์ราชา มวลประชาอยู่มาพ้นภัย', 'ขอดูแลคุ้มครองด้วยใจ นี่คือคำสัญญา']

f = open('เราจะทำตามสัญญา.txt', 'w+')
f.writelines(list(map(lambda l: l+'\n', lines)))
f.flush()
```

## Best Practices

context manager

```python
# เอา lyrics ใน never_gonna.txt ไปใส่ใน เราจะทำตามสัญญา.txt
with open('เราจะทำตามสัญญา.txt', 'a+') as my_file:
    with open('never_gonna.txt', 'r') as rick_rolled_file:
        buffer = rick_rolled_file.read()
        my_file.write(buffer)
```
