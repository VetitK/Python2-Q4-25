$$BMI = \frac{weight(kg)}{height^2 (m.)}$$

```python
class Person:
    def __init__(self, age, height, weight, name, mom_name):
        self.age = age
        self.height = height
        self.weight = weight
        self.name = name
        self.mom_name = mom_name
    
    def calculate_bmi(self):
        return self.weight / ((self.height/100)**2)
    
    def greeting(self):
        print("Hello")
```

```python
namo = Person(age=16, height=159, weight=48, name="Namo", mom_name="Runjuan")
kanplu = Person(age=13, height=160, weight=70, name="Kanplu", mom_name="JuanRun")
```

```python
kanplu.calculate_bmi()
```

```python
class CEO:
    def __init__(self, name, company, age, graduate):
        self.name = name
        self.company = company
        self.age = age
        self.graduate = graduate
        self.employee = []
        
    def hire_employee(self, employee_name):
        self.employee.append(employee_name)
        
    def fire_employee(self, employee_name):
        self.employee.remove(employee_name)
        
    def invest(self):
        print("Invest")
```

```python
class Toilet:
    def __init__(self, water=40, n_poop: float=0, heat=0, tissue=0):
        self.water = water
        self.n_poop = n_poop
        self.heat = heat
        self.tissue = tissue

    def poop(self, eww_poop: float):
        self.n_poop += eww_poop

    def flush(self):
        self.water = 0
        self.n_poop = 0
        return 'https://youtu.be/OqZmiDkAYPs?si=W6sHMkbRfGx992bI'

    def clean_butt(self, tissue_use_for_poop_tid_butt):
        if self.tissue <= 0:
            return 'Please use your underweaer'
        else:
            self.tissue -= tissue_use_for_poop_tid_butt
            return 'Your butt is clean like a baby!'
        
    def refill_tissue(self, amount):
        self.tissue + amount
```

```python
school_toilet = Toilet()
school_toilet.poop(5.5)
school_toilet.n_poop
school_toilet.poop(10)
school_toilet.n_poop
school_toilet.flush()
school_toilet.clean_butt(4)
```

```python
class Teacher:
    def __init__(self, teachers_name, homework: int, lessons, tests):
        self.teachers_name = teachers_name
        self.homework = homework
        self.lessons = lessons
        self.tests = tests

    def lesson(self):
        return f"{self.teachers_name} is teaching math"
    
    def assign_homework(self, n_homework: int=10):
        self.homework += n_homework
        if self.homework > 20:
            return "ตู้มมมมมมมมมมมมมม"
        return f"{self.teachers_name} assigned {n_homework} homework questions" 
```

```python
abdul = Teacher(teachers_name="Abdul", homework=0, lessons="Lamad", tests=0)
abdul.lesson()
abdul.assign_homework(5)
abdul.assign_homework(10)
abdul.assign_homework(5)
abdul.assign_homework(1)
```
