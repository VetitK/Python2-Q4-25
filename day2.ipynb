## Error Handling

```python
try:
    a = 2 + 2
    c = 1 + 1
    b = 3 / 0 
except:
    print(a)
    print(b)
    print(c)
```

Upgrade!

```python
try:
    a = 3 / 0 # ZeroDivisionError
    print(x) # NameError
    b = int("A") # ValueError
except NameError:
    print("Hello, World!")
```

LV3

```python
try:
    a = 3 / 0 # ZeroDivisionError
    print(x) # NameError
    b = int("A") # ValueError
except NameError:
    print("Hello, World!")
except ZeroDivisionError:
    print("Hello, Mars!")
except ValueError:
    print("Hello Jupyter!")
```

LV 3.5

```python
try:
    import x # ModuleNotFoundError
    b = int("A") # ValueError
    a = 3 / 0 # ZeroDivisionError
    print(x) # NameError
except (NameError, ZeroDivisionError):
    print("Hello, World!")
except ValueError:
    print("Hello Jupyter!")
except:
    print("อย่าหาทำ")
```

LV 4

```python
try:
    1+1
    # import x # ModuleNotFoundError
    # print(x) # NameError
    # b = int("A") # ValueError
    # a = 3 / 0 # ZeroDivisionError
except (NameError, ZeroDivisionError):
    print("Hello, World!")
except ValueError:
    print("Hello Jupyter!")
except:
    print("อย่าหาทำ")
finally:
    print("END")
```

LV 5

```python
try:
    print("NAMO")
    # import x # ModuleNotFoundError
    # print(x) # NameError
    # b = int("A") # ValueError
    # a = 3 / 0 # ZeroDivisionError
except (NameError, ZeroDivisionError):
    print("Hello, World!")
except ValueError:
    print("Hello Jupyter!")
except:
    print("อย่าหาทำ")
else:
    print("ELSE")
finally:
    print("END")
```
